```python
import tensorflow as tf
import numpy as np
```

```python
# load data
data = np.load("./data.npy") + 1
labels = np.load("./labels.npy") + 1 
m = max([np.max(data),np.max(labels)])
data/=m
labels/=m
data.shape,labels.shape
```

```python
data[0],data[1]
```

```python

# define model
model = tf.keras.Sequential()
model.add(tf.keras.layers.LSTM(256, activation='gelu', input_shape=(32, 10)))
model.add(tf.keras.layers.Dense(256,activation='gelu'))
model.add(tf.keras.layers.Dense(256,activation='gelu'))
model.add(tf.keras.layers.Dense(32,activation='gelu'))
model.compile(optimizer=tf.keras.optimizers.Adam(
    learning_rate=0.001,
    beta_1=0.9,
    beta_2=0.999,
    epsilon=1e-07,
    amsgrad=False,
    name='Adam',
), loss='mse')
```

```python


model.fit(x=data,y=labels,epochs=32,batch_size=32,shuffle=True)
```
